```python
import requests
import pandas as pd
```

```python
def get_warframe_market_items():
    url = "https://api.warframe.market/v1/items"
    
    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise an HTTPError for bad responses (4xx and 5xx)
        items_data = response.json()  # Parse the JSON response
        return items_data
    except requests.exceptions.HTTPError as http_err:
        print(f"HTTP error occurred: {http_err}")
    except Exception as err:
        print(f"An error occurred: {err}")
```

```python
def extract_relevant_info(items_data):
    items = items_data.get("payload", {}).get("items", [])
    
    filtered_items = items_data

    filtered_items = [
        {
            "url_name": item["url_name"],
            "item_name": item["item_name"]
        }
        for item in items
    ]

    return filtered_items
```

```python
items_data = get_warframe_market_items()
if items_data:
    filtered_items = extract_relevant_info(items_data)
    
    # Convert the filtered items to a DataFrame
    filtered_items_df = pd.DataFrame(filtered_items)
    
    # Create an empty DataFrame with the desired columns
    item_df = pd.DataFrame(columns=["url_name", "item_name"])
    
    # Append the filtered items DataFrame to the item_df
    item_df = pd.concat([item_df, filtered_items_df], ignore_index=True)
    
```

```python
def get_item_data():
    url = "https://api.warframe.market/v1/items/"
    try:
        response = requests.get(url + item_df.iloc[0].url_name + '/orders')
        response.raise_for_status()  # Raise an HTTPError for bad responses (4xx and 5xx)
        items_data = response.json()  # Parse the JSON response
    except requests.exceptions.HTTPError as http_err:
        print(f"HTTP error occurred: {http_err}")
    except Exception as err:
        print(f"An error occurred: {err}")
```
